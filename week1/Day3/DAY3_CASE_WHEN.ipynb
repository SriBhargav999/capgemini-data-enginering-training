```sql
-- Bonus Calculation Based on Department and Performance
SELECT *,
CASE
WHEN department='Finance' then 
CASE 
WHEN performance_rating='A' then '20% bonus'
WHEN performance_rating='B' then '15% bonus'
ELSE '10% bonus'
END
WHEN department='Engineering' then 
CASE 
WHEN performance_rating='A' then '18% bonus'
WHEN performance_rating='B' then '13% bonus'
ELSE '3% bonus'
END
ELSE '10% bonus'
END AS bonus
FROM employeeee
```

```sql
-- Categorizing Employees by Salary Range and Performance
SELECT *,
CASE 
WHEN salary>80000 and performance_rating='A' THEN 'High Performer'
WHEN salary>50000 and salary<80000 and performance_rating='B' THEN 'Mid Performer'
ELSE 'Low Performer'
END AS Performer
FROM employeeee
```
