### ROW NUMBER()

```sql
-- unique row number to all employees based on salary (highest first)
SELECT emp_name, salary, ROW_NUMBER() OVER (ORDER BY salary DESC) AS row_number
FROM employees1
```

```sql
-- row numbers to employees within each department based on salary descending. 
SELECT emp_name, department, salary,
ROW_NUMBER() OVER(PARTITION BY department ORDER BY salary DESC) AS row_number
FROM employees1
```

```sql
-- row numbers based on employee joining date (latest first)
SELECT emp_name, join_date,
ROW_NUMBER() OVER(ORDER BY join_date DESC) AS row_number
FROM employees1
```

```sql
-- row numbers within each department based on earliest joining date
SELECT emp_name, department, join_date,
ROW_NUMBER() OVER(PARTITION BY department ORDER BY join_date ) AS row_number
FROM employees1
```

### RANK()

```sql
-- Rank all employees based on salary (highest first)
SELECT emp_name, salary,
RANK() OVER(ORDER BY salary DESC) AS rank
FROM employees1
```

```sql
-- Rank employees within each department based on salary. 
SELECT emp_name, department, salary,
RANK() OVER(PARTITION BY department ORDER BY salary DESC) AS rank
FROM employees1
```

```sql
-- Rank employees based on joining date (latest gets rank 1). 
SELECT emp_name, join_date,
RANK() OVER(ORDER BY join_date DESC) AS rank
FROM employees1
```

```sql
-- Rank employees based on name alphabetically. 
SELECT emp_name,
RANK() OVER(ORDER BY emp_name) AS rank
FROM employees1
```

### DENSE_RANK()

```sql
-- dense rank to employees based on salary (highest first)
SELECT emp_name, salary,
DENSE_RANK() OVER(ORDER BY salary DESC) AS Dense_rank
FROM employees1
```

```sql
-- dense rank within each department based on salary
SELECT emp_name, department, salary,
DENSE_RANK() OVER(PARTITION BY department ORDER BY salary DESC) AS Dense_rank
FROM employees1
```

```sql
-- dense rank to employees based on salary (lowest first)
SELECT emp_name, salary,
DENSE_RANK() OVER(ORDER BY salary) AS Dense_rank
FROM employees1
```

```sql
-- dense rank within department based on joining date
SELECT emp_name, department, join_date,
DENSE_RANK() OVER(PARTITION BY department ORDER BY join_date) AS Dense_rank
FROM employees1
```
