### Phase 1: Data Preparation

```sql
%sql
CREATE OR REPLACE TABLE email_mapping AS
SELECT DISTINCT
    REGNO AS student_id,
    Student_Name,
    LOWER(TRIM(Email_Address)) AS primary_email,
    Branch
FROM my_csv_table
WHERE Email_Address IS NOT NULL;
```

```sql
%sql
-- Check for responses that don't match the Master List
SELECT 
    t2.Email_Address AS response_email,
    t2.Timestamp
FROM my_csv_table1 t2
LEFT JOIN email_mapping m ON t2.Email_Address = m.primary_email
WHERE m.student_id IS NULL;
```

### Phase 2: Core Analysis

```sql
-- Find students who have NOT submitted

SELECT 
    m.student_id,
    m.Student_Name,
    m.Branch,
    m.primary_email
FROM email_mapping m
LEFT JOIN my_csv_table1 t1 ON m.primary_email = t1.Email_Address
WHERE t1.Email_Address IS NULL
ORDER BY m.Branch, m.Student_Name;
```

```sql
-- Find valid submissions
SELECT 
    m.student_id,
    m.Student_Name,
    t1.Timestamp,
    t1.Completed_Status,
    t1.GitHub_Username
FROM email_mapping m
INNER JOIN my_csv_table1 t1 ON m.primary_email = t1.Email_Address
ORDER BY t1.Timestamp DESC;
```

```sql
-- Find invalid submissions
SELECT 
    t1.Email_Address AS Unknown_Email,
    t1.Timestamp,
    t1.GitHub_Username
FROM my_csv_table1 t1
LEFT JOIN email_mapping m ON t1.Email_Address = m.primary_email
WHERE m.student_id IS NULL;
```

### Phase 3: Duplicate Detection (Window Functions)

```sql
%sql
-- Use ROW_NUMBER() OVER (PARTITION BY student_id ORDER BY timestamp) to identify duplicates
CREATE OR REPLACE TABLE task_submissions_ranked AS
SELECT 
    m.student_id,
    m.Student_Name,
    t1.Email_Address,
    t1.Timestamp,
    ROW_NUMBER() OVER (
        PARTITION BY m.student_id 
        ORDER BY t1.Timestamp ASC
    ) as submission_rank
FROM email_mapping m
INNER JOIN my_csv_table1 t1 ON m.primary_email = t1.Email_Address;
```

```sql
-- Final View: Keeping only the first submission
CREATE OR REPLACE VIEW first_submissions_only AS
SELECT * FROM task_submissions_ranked WHERE submission_rank = 1;
```

### Phase 4: Advanced Insights

```sql
-- Count submissions per student.
SELECT 
    Student_Name, 
    COUNT(*) as attempt_count
FROM task_submissions_ranked
GROUP BY Student_Name
HAVING attempt_count > 1;
```

```sql
-- Find students using both emails
SELECT 
    student_id, 
    COUNT(DISTINCT Email_Address) as unique_emails_used
FROM (
    SELECT m.student_id, t1.Email_Address FROM my_csv_table1 t1 
    JOIN email_mapping m ON t1.Email_Address = m.primary_email
)
GROUP BY student_id
HAVING unique_emails_used > 1;
```

```sql
--  Generate final classification: Submitted, Not Submitted, Duplicate, Invalid

CREATE OR REPLACE TABLE final_student_report AS
SELECT 
    m.student_id,
    m.Student_Name,
    m.Branch,
    CASE 
        WHEN r.submission_rank = 1 THEN 'Submitted'
        WHEN r.submission_rank > 1 THEN 'Duplicate'
        WHEN m.student_id IS NOT NULL AND r.student_id IS NULL THEN 'Not Submitted'
        ELSE 'Invalid'
    END AS Final_Status
FROM email_mapping m
LEFT JOIN task_submissions_ranked r ON m.student_id = r.student_id;

SELECT * FROM final_student_report;
```
