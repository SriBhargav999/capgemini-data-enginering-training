### Identify nulls, duplicates, invalid values

```python
from pyspark.sql.functions import col
customers_null = customers.filter( col("customer_id").isNull() | col("city").isNull() | col("name").isNull())
cust_duplicates=customers.groupBy("customer_id").count().filter(col('count')>1)
customers_null.show()
cust_duplicates.show()
```

```python
cars_null=cars.filter( col("car_id").isNull() | col("brand").isNull() | col("model").isNull() | col("price").isNull())
cars_duplicates=cars.groupBy("car_id").count().filter(col('count')>1)
cars_invalid=cars.filter(col('price')<0)
cars_null.show()
cars_duplicates.show()
cars_invalid.show()
```

```python
dealers_null=dealers.filter( col("dealer_id").isNull() | col("name").isNull() | col("city").isNull())
dealers_duplicates=dealers.groupBy("dealer_id").count().filter(col('count')>1)
dealers_null.show()
dealers_duplicates.show()
```

```python
sales_null=sales.filter( col("sale_id").isNull() | col("customer_id").isNull() | col("car_id").isNull() | col("sale_date").isNull() | col("quantity").isNull())
sales_dupliactes=sales.groupBy("sale_id").count().filter(col('count')>1)
sales_null.show()
sales_dupliactes.show()
```

### Phase 4 – Transformation

```python
# Customer revenue
from pyspark.sql.functions import sum, col
customer_revenue= customers.join(valid_cars, "customer_id", "inner") \
  .groupBy("name") \
  .agg(sum("price").alias("revenue")) \
  .orderBy(col("revenue").desc())
customer_revenue.show()
```

```python
# Brand-wise sales
brand_sales=valid_cars\
    .groupBy ("brand")\
    .agg(count("quantity").alias("sales"))\
    .orderBy(col("sales").desc())
brand_sales.show()
```

```python
# City-wise revenue
city_revenue=customers.join(valid_cars,"customer_id","inner")\
    .groupBy("city")\
    .agg(sum("price").alias("revenue"))\
    .orderBy(col("revenue").desc())
city_revenue.show()
```

### Phase 5 – Dealer Analytics

```python
# Revenue per dealer
dealer_revenue=dealers.join(sales_dealer,"dealer_id","inner")\
    .join(valid_cars,"sale_id","inner")\
    .withColumn("revenue",col("price")*col("quantity"))\
    .groupBy("name")\
    .agg(sum(col("revenue")).alias("total_revenue"))\
    .orderBy(col("total_revenue").desc())
dealer_revenue.show()
```

```python
# Top dealers by revenue
dealer_revenue.limit(5).show()
```

```python
# Dealer-city contribution
dealer_revenue=dealers.join(sales_dealer,"dealer_id","inner")\
    .join(valid_cars,"sale_id","inner")\
    .withColumn("revenue",col("price")*col("quantity"))\
    .groupBy("name","city")\
    .agg(sum(col("revenue")).alias("total_revenue"))\
    .orderBy(col("total_revenue").desc())
dealer_revenue.show()
```

### Phase 6 – SQL Analysis

```python
# Top 3 customers per city
from pyspark.sql.window import Window
customer_revenue= customers.join(valid_cars, "customer_id", "inner") \
  .groupBy("name","city") \
  .agg(sum("price").alias("revenue")) \
  .orderBy(col("revenue").desc())
Window_spec=Window.partitionBy("city").orderBy(col("revenue").desc())
ranked=customer_revenue.withColumn("Rank" ,row_number().over(Window_spec))
top_3=ranked.filter(col("Rank")<=3)
top_3.show()
```

```python
# Monthly trends
sales=sales.withColumn("sale_date",to_date("sale_date"))\
    .withColumn("month",month("sale_date"))\
    .withColumn("year",year("sale_date"))\

monthly_trend = sales \
    .join(customers, "customer_id") \
    .join(valid_cars, "car_id") \
    .groupBy("year", "month","city") \
    .agg(sum("price").alias("total_revenue")) \
    .orderBy("year", "month")

monthly_trend.show()
```
